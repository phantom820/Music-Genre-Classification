### Low-level features
Low-level features pertain to content-based features. The chosen content-based features along with their representations are listed below.
- <b>Spectral centroid</b> - it indicates where the ”centre of mass” for a sound is located and is calculated as the weighted mean of the frequencies present in the sound (represented by mean over number of frames)
- <b>Spectral rolloff</b> - It indicates where the ”centre of mass” for a sound is located and is calculated as the weighted mean of the frequencies present in the sound(standard deviation over number of frames)
- <b>Chroma</b> - 12 bins representing the 12 distinct semitones (or chroma) of the musical octave. (mean over number of frames)
- <b>Zero crossing rate</b>the rate of sign-changes along a signal, i.e., the rate at which the signal changes from positive to negative or back (represented by mean over number of frames)
- <b>Mel frequency cepstral coefficients (MFCCs)</b> a small set of features (usually about 10–20) which concisely describe the overall shape of a spectral envelope. It models the characteristics of the human voice. (20 each represented by mean)
- <b>Root mean square energy</b>

A visualization of some of these features is shown below

```python
spectral_centroids = librosa.feature.spectral_centroid(song, sr=sr)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)# Normalising the spectral centroid for visualisation

# Normalising the spectral centroid for visualisation
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)

plt.figure(figsize=(14,5))
plt.title("Spectral centroid")
librosa.display.waveplot(song, sr=sr, alpha=0.4)
plt.plot(t,sklearn.preprocessing.minmax_scale(spectral_centroids,axis=0), color='r')

mfccs = librosa.feature.mfcc(song, sr=sr)
plt.figure(figsize=(14,5))
plt.title("MFCCs")
librosa.display.specshow(mfccs, sr=sr, x_axis='time')

hop_length = 512
chromagram = librosa.feature.chroma_stft(song, sr=sr, hop_length=hop_length)
plt.figure(figsize=(14, 5))
plt.title("Chroma")
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', hop_length=hop_length, cmap='coolwarm')

plt.show()
```

```python
def low_level_features(data):
    n=len(data)
    features_dataframe={"spectral_centroid":np.zeros(n),"spectral_rolloff":np.zeros(n),
                        "zero_crossing_rate":np.zeros(n),"chroma":np.zeros(n),"rms_mean":np.zeros(n),
                       "rms_std":np.zeros(n)}
    for index,row in data.iterrows(): 
        x,sr=librosa.load(row["filename"])
        #extract spectral centroid
        spectral_centroid=librosa.feature.spectral_centroid(x, sr=sr)
        features_dataframe["spectral_centroid"][index]=np.std(spectral_centroid)
        
        #extract spectral rolloff
        spectral_rolloff = librosa.feature.spectral_rolloff(x, sr=sr)
        features_dataframe["spectral_rolloff"][index]=np.std(spectral_rolloff)
        
        
        #extract chroma
        chroma=librosa.feature.chroma_stft(x, sr=sr)
        features_dataframe["chroma"][index]=np.mean(chroma)
        
#         for i in range(len(chroma)):
#             key="chroma"+str(i+1)
#             if key in features_dataframe.keys():
#                 features_dataframe[key][index]=(chroma[i])
#             else:
#                 features_dataframe[key]=np.zeros(n)
#                 features_dataframe[key][index]=(chroma[i])
        
        #extract zero crossing rate
        zcr=librosa.feature.zero_crossing_rate(x)
        features_dataframe["zero_crossing_rate"][index]=(np.mean(zcr))
        
        #extract root mean square energy
        rms=librosa.feature.rms(x)
        features_dataframe["rms_mean"][index]=np.mean(rms)
        features_dataframe["rms_std"][index]=np.std(rms)
        
        #extract mfccs (n mfccs)
        mfccs=librosa.feature.mfcc(x, sr=sr)
        mfccs=np.mean(mfccs,axis=1)
        
        for i in range(len(mfccs)):
            key="mfcc"+str(i+1)
            if key in features_dataframe.keys():
                features_dataframe[key][index]=(mfccs[i])
            else:
                features_dataframe[key]=np.zeros(n)
                features_dataframe[key][index]=(mfccs[i])
                
        #extract spectral contrast
        spectral_contrast = librosa.feature.spectral_contrast(x, sr=sr)
        spectral_contrast=np.mean(spectral_contrast,axis=1)
        
        for i in range(len(spectral_contrast)):
            key="sc"+str(i+1)
            if key in features_dataframe.keys():
                features_dataframe[key][index]=(spectral_contrast[i])
            else:
                features_dataframe[key]=np.zeros(n)
                features_dataframe[key][index]=(spectral_contrast[i])
                
    features_dataframe["label"]=data["genre id"].values
    features_dataframe=pd.DataFrame(features_dataframe)
    return features_dataframe
```

```python
start=time.time()
training_features=low_level_features(training_data)
training_features.to_csv("training_features.csv",index=False)
end=time.time()
print("Training set feature extraction duration :",end-start)
#training_features=pd.read_csv("training_features.csv")
# training_features=training_features.drop(columns=["chroma"+str(i) for i in range(1,13)])
training_features.head()
```

```python
start=time.time()
testing_features=low_level_features(testing_data)
testing_features.to_csv("testing_features.csv",index=False)
end=time.time()
print("Testing set feature extraction duration :",end-start)
#testing_features=pd.read_csv("testing_features.csv")
# testing_features=testing_features.drop(columns=["chroma"+str(i) for i in range(1,13)])
testing_features.head()
```

```python
x_train,y_train=training_features.loc[:,(training_features.columns!="label")].values,training_features["label"].values
x_test,y_test=testing_features.loc[:,(testing_features.columns!="label")].values,testing_features["label"].values
```

```python
x_train_norm=(x_train-np.mean(x_train,axis=0))/(np.std(x_train,axis=0)+1e-8)
x_test_norm=(x_test-np.mean(x_train,axis=0))/(np.std(x_train,axis=0)+1e-8)
```

## kNN classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(x_train_norm, y_train)
y=neigh.predict(x_train_norm)
y_predict=(neigh.predict(x_test_norm))
print("Train",np.count_nonzero(y==y_train)/len(y))
print("Test",np.count_nonzero(y_predict==y_test)/len(y_test))
```

### High-level features

A high-level feature would be a spectrogram obtained from the short time fourier transform of a given audio file. We may choose to compute the mel-spectrogram instead of the spectrogram. A mel-spectrogram is obtained by scaling a spectrogram to the mel-scale (pitch measure) this means that equally distances in pitch sound equally distant to the human ear. A visualization of the spectrogram and mel-spectrogram are shown below.

```python
audio_path='data/GTZAN/blues/blues.00002.wav'
x,sr=librosa.load(audio_path)
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))

chroma=librosa.feature.chroma_stft(x, sr=sr)
print(chroma.shape)

plt.figure()
plt.title("Spectrogram")
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')

plt.figure()
plt.title("Mel-spectrogram")
y, sr = librosa.load(audio_path)
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
mel_spect = librosa.power_to_db(X, ref=np.max)
librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time')

plt.show()
```
