```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import librosa
import librosa.display
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

## Load dataset

```python
data=pd.read_csv("../data/gtzan/features/deep_learning_features.csv")
data.head()
```

## Separate features and labels

```python
X,y=data.loc[:,(data.columns!="label")],data["label"].values
X.describe()
```

### Vectors to images

```python
x=np.zeros((1000,128,512,1))
x[:,:,:,0]=X.values.reshape((1000,128,512))
```

### Verify images

```python
plt.figure()
librosa.display.specshow(x[0,:,:,0], y_axis='mel', fmax=8000, x_axis='time')
plt.show()
```

## Model
- Convolutional neural network

### Train and test split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

### Feature normalization
#### Normalization parameters

```python
def normalization_parameters(v):
    mu = np.mean(v, axis=0)
    std = np.std(v, axis=0)
    return mu, std
```

#### Normalize features

```python
def normalize_features(v, mu, std):
    v = (v - mu) / (std + 1e-8)
    return v
```

```python
mu, std = normalization_parameters(x_train)
x_train_norm = normalize_features(x_train, mu, std)
x_test_norm = normalize_features(x_test,mu,std)
```

```python
mu,std = normalization_parameters(x)
x_norm = normalize_features(x,mu,std)
```

## Fitting  model

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Conv2D(16, (3, 3), activation='relu', input_shape=(128,64, 1)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(32, (3, 3), activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.02)))
    # model.add(layers.MaxPooling2D((2, 2)))
    # model.add(layers.Conv2D(32, (3, 3), activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.02)))
    # model.add(layers.MaxPooling2D((2, 2)))
    # model.add(layers.Conv2D(64, (3, 3), activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.02)))
    model.add(layers.Flatten())
    model.add(layers.Dense(32, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.02)))
    model.add(layers.Dense(10))
    
    return model
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history = model.fit(x_train_norm[:,:,:64,:], y_train, epochs=5, 
                    validation_data=(x_test_norm[:,:,:64,:], y_test))
```

```python
print(history.history)
```

```python
def fit_model():
    epochs_data=[]
    kf = KFold(n_splits=10,shuffle=True)
    for train_index, test_index in kf.split(x_norm):
        X_train, X_test = x_norm[train_index], x_norm[test_index]
        Y_train, Y_test = y[train_index], y[test_index]
        model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
        history = model.fit(X_train[:,:,:64,:], Y_train, epochs=5, 
                    validation_data=(X_test[:,:,:64,:], Y_test),verbose=0)
        epochs_data.append(history.history["val_accuracy"][-1])
        print(epochs_data)
    return epochs_data

kfold_data=fit_model()
print(kfold_data)
```
