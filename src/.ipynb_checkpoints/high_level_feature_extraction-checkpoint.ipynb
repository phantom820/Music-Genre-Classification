### High-level features
High-level features pertain to abstract music features. The chosen high level feature is the
- Spectrogram

### Spectrogram computation

```python
def compute_spectrogram(x,sr):
    mel_spect = librosa.feature.melspectrogram(y=x, sr=sr, n_fft=2048, hop_length=1024)
    mel_spect = librosa.power_to_db(mel_spect, ref=np.max)[:,:512]
    return mel_spect
```

```python
genre_names={0: "blues", 1: "classical",2: "country",3: "disco",
 4: "hiphop", 5: "jazz",6: "metal",7: "pop", 8: "reggae",9: "rock"}
```

### Spectrogram visualization 

```python
fig,ax=plt.subplots(5,2,sharex=True,figsize=(15,17))
for i in range(5):
    for j in range(2):
        sample=data.loc[data["genre id"]==i+5*j].iloc[0].filename
        x,sr = librosa.load(sample)
        mel_spect=compute_spectrogram(x,sr)
        image=librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time',ax=ax[i,j])
        ax[i,j].set(title=genre_names[i+5*j].capitalize())
```

### Feature extraction step

```python
def high_level_features(data):
    n=128*512
    features_matrix=np.zeros((1000,n))
    for index,row in data.iterrows():
        x, sr = librosa.load(row["filename"])
        mel_spect=compute_spectrogram(x,sr).flatten()
        features_matrix[index,0:min(n,len(mel_spect))]=mel_spect[0:min(n,len(mel_spect))]
    features_dataframe=pd.DataFrame(features_matrix)
    features_dataframe["label"]=data["genre id"]
    return features_dataframe
```

```python
start=time.time()
features=high_level_features(data)
print("Feature extraction duration",time.time()-start)
features.head()
```

```python
features.to_csv("../data/gtzan/features/deep_learning_features.csv",index=False)
```

### Vector to image

```python
def vector_to_image(vector):
    image=vector.reshape((128,512))
    return image
```

### Spectrogram from vector

```python
f=features.drop(columns="label")
image=vector_to_image(f.loc[90].values)
librosa.display.specshow(image, y_axis='mel', fmax=8000, x_axis='time')
plt.show()
```
